Функция `description` на вход получает имя таблицы и количество наблюдений, которые нужно вывести для ознакомления. На выходе вернёт случайно выбранные наблюдения, таблицу с постолбцовым описанием данных, диапазон индексов и общее число записей таблицы.

```python
def description(data, ns): 
    descripted_data = pd.DataFrame({'Наблюдений': data.count(),
                                    'Уникальных значений': data.nunique(),
                                    'Пропусков': data.isna().sum(), 
                                    'Доля пропусков': round(data.isna().sum() / len(data), 3), 
                                    'Тип данных': data.dtypes})
    display(data.sample(ns))
    display(descripted_data)
    print(f'Диапазон индексов таблицы: {data.index.min()} - {data.index.max()}')
    print(f'Количестов записей в таблице: {len(data)}')
    print(f'Количество строк-дубликатов: {data.duplicated().sum()}')
    print('Доля строк-дубликатов дубликатов:{0:.2f}'.format(data.duplicated().sum()/ len(data)))
    print(f'Количество столбцов: {len(data.columns)}')
    return
```

Функция `content` на вход получает название столбца, возвращает количество и долю пропусков. Для столбцов с количеством уникальных значений не более 10 вернёт число наблюдений по каждому значению. Для столбцов с 2-3 категориями рисует круговую диаграмму, Если значений не больше 10 столбцовую. Для числовых параметров ящик с усами.

```python
def content(col_name):
    print(f"Количество пропусков: {data[col_name].isna().sum()}")
    print(f"Доля пропусков: {round(data[col_name].isna().sum() / len(data), 3)}")
    if data[col_name].nunique() <= 3:
        print(f'Количество наблюдений:\n{data[col_name].value_counts()}')
        plt.figure(figsize=(4,4))
        plt.pie(data[col_name].value_counts(), autopct='%1.2f%%', startangle=90, 
                colors = sns.color_palette('pastel')[ 0:3 ])
        plt.legend(bbox_to_anchor=(1, 0, 0.5, 1), labels = data[col_name].value_counts().index)
    elif data[col_name].nunique() <= 10:
        print(f'Количество наблюдений:\n{data[col_name].value_counts()}')
        plt.figure(figsize=(6, 4))
        ax = sns.barplot(x=col_name, y='count', data = (data.groupby(col_name)
                                                          .agg(count=(col_name,'count'))
                                                          .reset_index()))
        plt.title(col_dict.get(col_name))
        plt.xlabel(col_dict.get(col_name))
        plt.ylabel('Количество наблюдений')
        for p in ax.patches:
            percentage = '{:.1f}%'.format(100 * p.get_height()/len(data[col_name]))
            x = p.get_x()
            y = p.get_y() + p.get_height()
            ax.annotate(percentage, (x, y))
        plt.show()
    else:
        print(data[col_name].describe())
        plt.figure(figsize=(4,4))
        ax = sns.boxplot(y = data[col_name])
        ax.set_ylabel(col_dict.get(col_name))
        ax.set_title(col_dict.get(col_name))
        plt.show()
    return
```

***Функции для отрисовки графиков***

Функция `draw_bars` на вход получает названия колонок для группировки и данные для построения, возвращает сгруппированную таблицу с долей второго признака внутри первого и от общего, столбчатую диаграмму(barplot) с разбивкой по второму признаку. 

```python
def draw_bars(col_one, col_two, data = data):        
        group_data = (data.groupby([col_one, col_two])
                      .agg(count=(col_one,'count')))
        group_data = data.groupby([col_one, col_two]).agg(count=(col_one,'count'))
        products_pcts = group_data.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum()))
        group_data['%_in_group']= round(products_pcts['count'], 2)
        group_data = group_data.reset_index()
        group_data['%_of_all'] = round(group_data['count']/data[col_two].count()*100, 1)
        display(group_data)
        plt.figure(figsize=(8, 3))
        ax = sns.barplot(x=col_one, y='count', data = group_data, hue = col_two)
        plt.title(f'{col_dict.get(col_one)} в разрезе {col_dict.get(col_two)}')
        plt.xlabel(col_dict.get(col_one))
        plt.ylabel('Количество наблюдений')
        plt.legend(loc='upper right')
        for p in ax.patches:
            count = '{:.0f}'.format(p.get_height())
            x = p.get_x()
            y = p.get_y()+ p.get_height()
            ax.annotate(count, (x, y))
        plt.show()
```

Функция `draw_boxplots` на вход получает целевую колонку(target_col), по которой будет происходить разбиение, список колонок по которым будут построены графики(col_list) и таблицу (data). По умолчанию из данных исключены выбросы баланса и заработной платы.

```python
def draw_boxplots(target_col, col_list, data = data.query('big_balance == 0 & big_salary == 0')):
    plt.figure(figsize=(20, 20))
    j = 0
    for col in col_list:
            j+=1
            sns.boxplot(y = col, x = target_col, data = data,
            ax=plt.subplot(3, 3, j))
            plt.title(f'{col_dict.get(col)} в разрезе\n{col_dict.get(target_col)}')
            plt.xlabel(col_dict.get(target_col))
            plt.ylabel(col_dict.get(col))
```

Функция `count_blowout` для построения сводной таблицы с количеством клиентов с большим балансом на счёте и большой заработной платой. На основе сводной строит круговые диаграммы.

```python
def count_blowout(target_col, data = data.query('big_balance == 1 or big_salary == 1')):
    group_data = (data.groupby(target_col)
         .agg(big_balance=('big_balance','sum'),
              big_salary = ('big_salary','sum'))
         .sort_values(by = 'big_balance', ascending = False))
    display(group_data)
    plt.figure(figsize=(10,4))
    j = 0
    for col in group_data.columns:
        j+=1
        ax=plt.subplot(1, 2, j)
        ax.figsize=(3,3)
        plt.pie(group_data[col], autopct='%1.1f%%', startangle=90)
        plt.title(f'Доля клиентов в категории\n{col_dict.get(col)}\n по признаку {col_dict.get(target_col)}')
        plt.legend(bbox_to_anchor=(0.7, 0, 0.5, 1), labels = group_data.index)
```

## Проверка статистических гипотез

Функция `draw_hists` для отрисовки распределений по необходимым признакам, base_col столбец, по которому ведется сравнение, col- критерий сравнения, data- данные для отрисовки, bins - количество корзин, по умолчанию определяется автоматически.

```python
def draw_hists(base_col, col, data = data, bins = 'auto'):
    plt.figure(figsize =(8,4))
    sns.histplot( data = data, x = data[base_col], hue = col, palette = 'BuPu',
                 stat = 'density',alpha = 0.4, kde = True, bins = bins, common_norm = False)
    plt.title(f'Распределение по признаку {col_dict.get(col)}')
    plt.xlabel(col_dict.get(base_col))
    plt.ylabel('плотность')
       
```

Функция `mannwhitneyu` проводит проверку двух выборок, используя тест Манна-Уитни, на вход получает две выборки(slice_one, slice_two) и признак, по которому проводится сравнение,также на вход передаётся критический уровень значимости, по умолчанию равен 0.05. Возвращает p_value и вердикт о возможности отвергнуть нулевую гипотезу.

```python
def mannwhitneyu(slice_one, slice_two, target_col, alpha =0.05):
    results = st.mannwhitneyu(
        slice_one[target_col], 
        slice_two[target_col], 
        alternative = 'two-sided')
    print('p_value:', results.pvalue)
    if results.pvalue < alpha:
        print('Отвергаем нулевую гипотезу')
    else:
        print('Не получилось отвергнуть нулевую гипотезу')
```

### Проверка гипотезы  различия возраста между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним.
H0: Возрасты клиентов использующих один продукт и два продукта одинаковы

H1: Возрасты клиентов использующих один продукт и два продукта отличаются

#### Проверка распределений

```python
draw_hists('age','products', data = data.query('products in (1,2)'))
```

Для проверки гипотезы будем использовать тест Манна-Уитни,так как видим, что распределение не нормально и знаем, что количество наблюдений достаточно велико. Критический уровень значимости p_value примем равным 0.05, будем использовать двусторонний метод проверки.

#### Результат теста

```python
mannwhitneyu(data.query('products == 1'), data.query('products == 2'), 'age')
```

Значение p_value чуть больше 0.05, мы не можем отвергнуть нулевую гипотезу, значит нет статистически значимых различий в возрастах клиентов использующих один продукт и два.

### Проверка гипотезы различия уровня заработных плат у клиентов использующих кредитную карту и не использующих

H0: Уровни заработных плат у клиентов использующих кредитную карту и неиспользующих одинаковы

H1: Уровни заработных плат у клиентов использующих кредитную карту и неиспользующих отличаются

### Сегменты клиентов 

Исходя из результатов анализа можно сформировать 6 сегментов клиентов:

    1: клиенты использующие 1-2 продукта банка
    2: клиенты использующие 3 продукта
    3: клиенты пользующиеся 4-5 продуктов
    4: клиенты с высокой заработной платой
    5: клиенты с крупным балансом
    6: кредитные клиенты, у которых отсутствует информация о балансе
Сформируем портреты клиентов и рекомендации для них. Будем использовать функцию `product_group` для агрегации параметров.

```python
def product_group(col, data):
    display(data.pivot_table(index=col,
                          aggfunc={
                            'score': 'median',
                            'age': 'median',
                            'balance': 'median',
                            'equity': 'median',
                            'estimated_salary': 'median',
                            'churn': 'mean',
                            'credit_card': 'mean',
                            'man': 'mean',
                            'last_activity':'mean',
                            'age_category':'median',
                            'category_salary':'median'})\
             .reset_index())
```
